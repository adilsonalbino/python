```python
# Importando bibliotecas
import pandas as pd
from google.colab import files

# Carregar a planilha original
# Faça o upload do arquivo para o Google Colab
uploaded = files.upload()

# Escolha o arquivo a ser carregado
planilha_uploaded = list(uploaded.keys())[0]

# Carregue o arquivo com o Pandas e mantenha o índice original
planilha_original = pd.read_excel(planilha_uploaded, index_col=None)

# Converter números de série do Excel (base 01/01/1900)
planilha_original['Data de Criação'] = pd.to_datetime('1900-01-01') + pd.to_timedelta(planilha_original['Data de Criação'] - 2, unit='D')
planilha_original['Data de Finalização'] = pd.to_datetime('1900-01-01') + pd.to_timedelta(planilha_original['Data de Finalização'] - 2, unit='D')


planilha_original.head()
```

```python
# Ordem correta das colunas
ordem_correta = [
    'Data de Criação', 'Nº Chamado', 'Nome do Cliente', 'Assunto', 'Descrição',
    'Localizador GDS', 'Lista de Procedimentos', 'Tipos de Negação', 'Lista de Produtos',
    'Emergencial - Unidade de Atendimento', 'Nome completo do Passageiro',
    'Matricula do Cliente', 'Centro de Custo do Cliente', 'Nome Completo do Operador',
    'Email do Solicitante', 'Data de Finalização', 'Valor'
]

# Adicionar colunas que estão na ordem correta, mas não na planilha original
for col in ordem_correta:
    if col not in planilha_original.columns:
        planilha_original[col] = ''

# Manter apenas as colunas na ordem correta
planilha_ordenada = planilha_original[ordem_correta]

planilha_ordenada.head()
```

```python
# Salvar a nova planilha corrigida no diretório de destino
caminho_planilha_ordenada = 'palnilha_ordenada.xlsx'
planilha_ordenada.to_excel(caminho_planilha_ordenada, na_rep='', index=False)

# Baixar o arquivo para o seu computador
files.download(caminho_planilha_ordenada)

# Informar que o processo foi concluído
print("Planilha ordenada baixada com sucesso!")
```
